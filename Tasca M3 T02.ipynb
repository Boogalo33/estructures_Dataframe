```python
#Ejercicio1

import pandas as pd
import numpy as np

#Subimos el archivo en jupyeter y luego los cargamos para poder trabajar con el.
df = pd.read_csv('DelayedFlights.csv')

# Mostramos las primeras filas del DataFrame
print(df.head())

# Mostramos información sobre el DataFrame para obtener detalles sobre las variables
print(df.info())

#Seleccionamos los encabezados de las columnas del excel.
columnas_relevantes = ['Year', 'Month', 'DayOfWeek', 'DepTime', 'ArrTime', 'UniqueCarrier', 'Origin', 'Dest', 'Distance', 'ArrDelay']
df_reducido = df[columnas_relevantes]


# Reducimos la dimensión de manera aleatoria a 200.000 registros.
df_reducido = df_reducido.sample(n=200000, random_state=42)

# Mostramos las primeras filas del Dataframa reducido.
print(df_reducido.head())
```

```python
#Ejercicio 2
#Podemos analizar la relación entre el tiempo de vuelo y el retraso, retrasos de salida y llegada e identificar
#las tendencias mensuales y semanales
df.describe()

print("Media de retraso por llegada:", df['ArrDelay'].mean())
print("Desv.standard de los retrasos de llegada:", df['ArrDelay'].std())

# Troba quantes dades faltants hi ha per columna
df.isnull().sum()

# Suponemos que "Airtime" Representa el tiempo de vuelos en minutos
df['Velocidadv media'] = df['Distance'] / df['AirTime']


# Clasificamos que un vuelo se considera tarde si la llegada es más de 20minutos a su destino
df['Llegada con Retraso'] = df['ArrDelay'] > 20

# Agrupamos por compañia y sumamos los retrasos
retrasos_por_aerolinia = df.groupby('UniqueCarrier')['ArrDelay'].sum().sort_values(ascending=False)


# Suponemos que "Distance" representa la distancia en Km
# Buscamos las rutas más largas
vuelos_mas_largos = df.groupby(['Origin', 'Dest'])['Distance'].sum().sort_values(ascending=False)

# La ruta con más retrasos
vuelos_mas_retrasados = df.groupby(['Origin', 'Dest'])['ArrDelay'].sum().sort_values(ascending=False)

```
